## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df =df.fillna(method='ffill')
````

````python
print(df.isnull().values.sum())
````

````python
# no. of Listings by Room_type

room_count=df["room_type"].value_counts()
room_count
````

````python
#average price by neighbourhood_group

avg_price_by_neighborhood = df.groupby(['neighbourhood_group'])['price'].mean().reset_index()
avg_price_by_neighborhood
````

````python
#grouping and checking average price, availability by room type

room_type_analysis = df.groupby('room_type').agg({'price': 'mean', 'availability_365': 'mean'}).reset_index()
room_type_analysis
````

````python
#Checking avg. nights and availability by Room type

minimum_nights_analysis = df.groupby('room_type').agg({'minimum_nights':'mean','availability_365': 'mean'}).reset_index()
minimum_nights_analysis
````

````python
#Number of Neighbourhoods

neighbourhood_count =  df['neighbourhood'].nunique()
print("Number of Neighbourhood:", neighbourhood_count)
````

````python
# Number of Neighbourhood groups

neighbourhood_group_count = df['neighbourhood_group'].nunique()
print("Number of Neighbourhood Group:", neighbourhood_group_count)
````

````python
#Popular hosts by No. of listings

popular_hosts= df.groupby('host_name')['calculated_host_listings_count'].size().nlargest(5).reset_index()
popular_hosts
````

````python
# Popular neighbourhoods by total number of reviews

popular_neighbourhood = df.groupby('neighbourhood').agg({'number_of_reviews':'mean','reviews_per_month':'mean'}).nlargest(5, ['number_of_reviews', 'reviews_per_month']).reset_index()
popular_neighbourhood
````

````python
# availability by room_type

room_type_availability = df.groupby('room_type').agg({'availability_365':'mean'}).reset_index()
room_type_availability
````

````python
df['last_review'] = pd.to_datetime(df['last_review'])
df['month'] = df['last_review'].dt.month.copy()
total_bookings_by_month_neighbourhood_group = df.groupby(['month', 'neighbourhood_group']).size().reset_index(name='total_bookings')
total_bookings_by_month_neighbourhood_group
````

````python
room_percentage = (room_count / len(df)) * 100
room_percentage
````

````python
neighbourhood_group_availability = df.groupby('neighbourhood_group').agg({'availability_365':'mean'}).reset_index()
neighbourhood_group_availability
````

````python
neighbourhood_availability = df.groupby(['neighbourhood', 'neighbourhood_group']).agg({'availability_365':'mean', 'price': 'mean'}).nlargest(5,['availability_365', 'price']).reset_index()
neighbourhood_availability
````
